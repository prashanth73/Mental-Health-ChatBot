```python
model = AutoModelForCausalLM.from_pretrained(
    base_model,
    quantization_config=bnb_config,
    device_map="auto",
    attn_implementation=attn_implementation
)
tokenizer = AutoTokenizer.from_pretrained(base_model, trust_remote_code=True)
```

```python
def find_all_linear_names(model):
    cls = bnb.nn.Linear4bit
    lora_module_names = set()
    for name, module in model.named_modules():
        if isinstance(module, cls):
            names = name.split('.')
            lora_module_names.add(names[0] if len(names) == 1 else names[-1])
    lora_module_names.discard('lm_head')  # Exclude lm_head for 16-bit
    return list(lora_module_names)

modules = find_all_linear_names(model)
modules
```

```python
peft_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=modules
)

tokenizer.chat_template = None

model, tokenizer = setup_chat_format(model, tokenizer)
model = get_peft_model(model, peft_config)
```

```python
from datasets import load_dataset

dataset = load_dataset("csv", data_files="/content/counsel_data.csv")
dataset = dataset['train'].shuffle(seed=42).select(range(10000))

def format_chat_template(row):
    row_json = [{"role": "user", "content": row["input"]},
                {"role": "assistant", "content": row["output"]}]
    row["text"] = tokenizer.apply_chat_template(row_json, tokenize=False)
    #row["text"] = response_tokenizer.apply_chat_template(row_json, tokenize=False)
    return row

dataset = dataset.map(format_chat_template, num_proc=4)

dataset = dataset.train_test_split(test_size=0.05)
```

```python
peft_model_id = "Prashanth74/Gemma_2b_Response_Generation_Model_MentalHealth_Chatbot"
peft_config = PeftConfig.from_pretrained(peft_model_id)
response_tokenizer = AutoTokenizer.from_pretrained(peft_config.base_model_name_or_path)

new_tokens = ["<extra_token_1>", "<extra_token_2>"]
response_tokenizer.add_tokens(new_tokens)
#print("Tokenizer size after adding tokens:", len(response_tokenizer))  # Should now be 256002

base_model = AutoModelForCausalLM.from_pretrained(
    peft_config.base_model_name_or_path,
    device_map="auto"
)

base_model.resize_token_embeddings(len(response_tokenizer))
response_model = PeftModel.from_pretrained(base_model, peft_model_id)
```

```python
generated_responses = []
reference_responses = []

test_set = dataset["test"]

for row in tqdm(test_set):
    # Create chat prompt
    messages = [
      {"role": "user", "content": row["input"]}
    ]
    prompt = response_tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = response_tokenizer(prompt, return_tensors="pt", padding=True, truncation=True).to("cuda")

    with torch.no_grad():
        outputs = response_model.generate(
            **inputs,
            max_length=800,
            top_k=50,
            do_sample=True,
            temperature=0.3,
            no_repeat_ngram_size=3
        )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True).split("model")[-1].strip()

    generated_responses.append(response)
    reference_responses.append(row["output"])
```

```python
P, R, F1 = score(generated_responses, reference_responses, lang="en", model_type="bert-base-uncased")

print(f"Average BERTScore Precision:  {P.mean().item():.4f}")
print(f"Average BERTScore Recall:     {R.mean().item():.4f}")
print(f"Average BERTScore F1 Score:   {F1.mean().item():.4f}")
```
